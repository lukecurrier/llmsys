```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import os
import shutil

# 'a' = American English, ğŸ‡¬'b' => British English
pipeline = KPipeline(lang_code='a', repo_id='hexgrad/Kokoro-82M') # <= make sure lang_code matches voice
```

```python
VOICE = 'af_heart'
```

```python
text = ""
article_name = "attentionisallyouneed"

with open(f'conversions/{article_name}.md') as f:
    text += f.read()
```

```python
cut_text = text[:500]
print(cut_text)
```

```python
generator = pipeline(
    cut_text, 
    voice=VOICE,
    speed=1, 
    split_pattern=r'\n+'
)
```

```python
output_dir = "sound_files"

if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

os.makedirs(output_dir)

for i, (gs, ps, audio) in enumerate(generator):
    print(i)  
    print(gs) 
    print(ps) 
    display(Audio(data=audio, rate=24000, autoplay=i==0))

    output_path = os.path.join(output_dir, f"{i}.wav")
    sf.write(output_path, audio, 24000)
```
